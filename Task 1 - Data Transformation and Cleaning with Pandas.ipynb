### Step 0
Load the energy data from the file `Energy Indicators.xls`, which is a list of indicators of [energy supply and renewable electricity production](Energy%20Indicators.xls) from the [United Nations](http://unstats.un.org/unsd/environment/excel_file_tables/2013/Energy%20Indicators.xls) for the year 2013, and should be put into a DataFrame with the variable name of **energy**.

Keep in mind that this is an Excel file, and not a comma separated values file. Also, make sure to exclude the footer and header information from the datafile. The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:

`['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']`

Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule). For all countries which have missing data (e.g. data with "...") make sure this is reflected as `np.NaN` values.

Rename the following list of countries (for use in later questions):

```"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"```

There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 

e.g.

`'Switzerland17'` should be `'Switzerland'`.

````python
import pandas as pd
import numpy as np
````

````python
## .iloc[8:,2:] : Name of columns are row 9; the date starts at the 17th row; so we remove the first 8.
# the country column can be found twice plus a null column; we remove the first 2 cols.
## .reset_index().iloc[:,1:] : The index after the first step starts at 8 so reset it and remove the previous index column
energy=pd.read_excel('./data/Energy Indicators.xls',nrows=235,header=[9],index_col=None).iloc[8:,2:].reset_index().iloc[:,1:]

# We Change the column names
energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

# We change petajoules to gigajoules
energy['Energy Supply']=energy['Energy Supply'].apply(lambda x: x*(10**6))

#Since all cols (except for countries) hold numeric values thus any non numeric value should be null this is done using regex
energy.iloc[:,1:]=energy.iloc[:,1:].replace(regex=r'\D',value=np.nan)

# Rename the following list of countries (for use in later questions):
# "Republic of Korea": "South Korea",
# "United States of America": "United States",
# "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
# "China, Hong Kong Special Administrative Region": "Hong Kong"
values=["South Korea","United States","United Kingdom","Hong Kong"]
to_replaces=[
    "Republic of Korea"
    ,"United States of America"
    ,"United Kingdom of Great Britain and Northern Ireland"
    ,"China, Hong Kong Special Administrative Region"]
````

````python
#Testing line 17 to 26 BEFORE
print(energy[energy['Country'].isin(values)])

# Investigate unclean country names (contains nums or paranthesis) BEFORE
unclean_bol=energy['Country'].str.extract(r'(\d|\()').fillna(0)!=0
unclean_ser=energy[unclean_bol.iloc[:,0]]['Country']
print(unclean_ser)
````

````python
# Function to remove trailing unwanted chars
def clean_country(text):
    ret_str=''
    for ind in range(len(text)):
        if text[ind] == '(' or text[ind].isdigit():
            if ret_str[-1]==' ':
                ret_str=ret_str[:-1]
            break
        ret_str=ret_str+text[ind]
    return ret_str
````

````python
# Cleaning the countries names by applying the clean_country function to each entry
energy['Country']=energy['Country'].apply(clean_country)

# Here we apply the updates
energy['Country']=energy['Country'].replace(to_replace=to_replaces,value=values )   

energy
````

````python
#Testing line 17 to 26 AFTER
print(energy[energy['Country'].isin(values)])

# Investigate unclean country names (contains nums or paranthesis) AFTER
unclean_bol=energy['Country'].str.extract(r'(\d|\()').fillna(0)!=0
unclean_ser=energy[unclean_bol.iloc[:,0]]['Country']
print(unclean_ser)
````

### Step 3
Join the three datasets: GDP, Energy, and ScimEn into a new dataset (using the intersection of country names). Use only the last 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15). 

The index of this DataFrame should be the name of the country, and the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',

````python
# A list of year columns needed; to be used during merging process
year_list=[str(i) for i in range(2006,2016)]

# Merging 'ScimEn' and 'energy' together and assigning it to Result dataset
Result = pd.merge(ScimEn.sort_values(by='Rank')
                ,energy
                ,on='Country'
                ,how='inner')

# Merging the temp Result and 'GDP' together and updating result dataset
Result = pd.merge(Result
                 ,GDP[['Country']+year_list]
                 ,on='Country'
                 ,how='inner')

# Sorting the Result dataset based on the rank and setting the 'Country' column as the index
Result.sort_values(by='Rank',inplace=True)
Result.set_index(keys='Country',inplace=True)

# Checking the dimensions of the Result dataset
print("Result dimensions: "+str(Result.shape))

# Getting the top 15 countries
Result_top_15=Result.iloc[:15,:]

# Checking the dimensions of the Result_top_15 dataset
print("Result_top_15 dimensions: "+str(Result_top_15.shape))

# Checking the results
Result_top_15
````

````python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
continent_df=pd.DataFrame(data=list(zip(ContinentDict.keys(),ContinentDict.values())),columns=['Country','Continent'])

# Merging 'continent_df' and 'Result_top_15' together and assigning it to 'Result_top_15' dataset
Result_top_15 = pd.merge(continent_df
                ,Result_top_15
                ,on='Country'
                ,how='left')
````
